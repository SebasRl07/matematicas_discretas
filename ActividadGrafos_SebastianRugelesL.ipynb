GRAFO PONDERADO

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear Grafo
G= nx.Graph()

# Añadir vértices
G.add_nodes_from(['NY', 'SF', 'BO', 'CH', 'LA'])

# Añadir aristas con pesos
edges_with_weights = ([('NY','BO',20),('NY','SF',80),('NY','CH',30),('BO','CH',25),('BO','LA',60),('LA','CH',50),('LA','SF',40)])
G.add_weighted_edges_from(edges_with_weights)

# Crear tamaño específico
plt.figure(figsize=(4, 3))

# Dibujar el grafo
Pos = nx.spring_layout(G, seed=30)
nx.draw(G, Pos, with_labels=True, font_weight='bold', node_color='skyblue', font_size=11, node_size=700)

# Añadir etiquetas de pesos a las aristas
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, Pos, edge_labels=edge_labels)

# Mostrar el grafo
plt.show()
```

CAMINOS

```python
# Encontrar un camino entre dos nodos
path = nx.shortest_path(G, source='NY', target='LA')
print(f"Un camino entre NY y LA: {path}")

# Encontrar un recorrido (en este caso, un camino sin aristas repetidas)
print(f"Un recorrido entre NY y LA: {path}")

# Encontrar circuitos (un camino que empiece y termine en el mismo nodo)
cycles = list(nx.find_cycle(G, orientation='ignore'))
if cycles:
    print(f"Circuitos: {cycles}")
else:
    print("Circuitos: No hay circuitos en este grafo.")

# Encontrar caminos simples (un camino sin nodos repetidos)
print(f"Un camino simple entre NY y LA: {path}")

# Encontrar ciclos (un ciclo es una secuencia de nodos que empieza y termina en el mismo nodo)
simple_cycles = list(nx.simple_cycles(G.to_directed()))  # Convertir a grafo dirigido para encontrar ciclos
if simple_cycles:
    print(f"Ciclos: {simple_cycles}")
else:
    print("Ciclos: No hay ciclos en este grafo.")
```
